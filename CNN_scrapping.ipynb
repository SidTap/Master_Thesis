```python
from bs4 import BeautifulSoup
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the HTML content from the file
with open("C:\\Users\\siddd\\OneDrive\\Desktop\\srh\\master_Thesis\\code\\news_articles\\CNN\\bollywood_CNN.txt", 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the relevant div elements with the 'data-open-link' attribute
article_cards = soup.find_all('div', attrs={'data-open-link': True})

# Initialize a list to hold the URLs
urls = []

# Extract 'data-open-link' values (URLs) and print them
for card in article_cards:
    urls.append(card['data-open-link'])
    print(card['data-open-link'])

# Check number of extracted URLs
print(f"Total articles found: {len(urls)}")
```

```python
# Create a list to hold all article data
articles = []


def extract_content(url):
    try:
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
        response.raise_for_status()  # Ensure the request succeeded
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Extract heading
        heading = soup.find('h1').get_text(strip=True) if soup.find('h1') else 'No Heading Found'
        
        # Try extracting date with class 'update-time'
        date = soup.find('span', {'class': 'update-time'}).get_text(strip=True) if soup.find('span', {'class': 'update-time'}) else None
        
        # If the specific class didn't work, use regex to find a date pattern
        if not date:
            date_text = soup.get_text()
            date_match = re.search(r'\b(January|February|March|April|May|June|July|August|September|October|November|December) \d{1,2}, \d{4}\b', date_text)
            date = date_match.group(0) if date_match else 'No Date Found'
        
        # Extract content
        content = ' '.join([p.get_text(strip=True) for p in soup.find_all('p')])
        
        # Append the extracted data to the articles list
        articles.append({
            'URL': url,
            'Heading': heading,
            'Date': date,
            'Content': content
        })
    except Exception as e:
        print(f"Failed to process {url}: {str(e)}")


# Process each URL
for url in urls:
    extract_content(url)

# Save the articles to a CSV file
df = pd.DataFrame(articles)
df.to_csv('Bollywood_CNN.csv', index=False)
print("Data has been successfully extracted and saved to CSV.")
df
```
